As shown in the figure above, the perceptron has a few inputs and a single output. A perceptron is traditionally assumed to have a single output though neurons with multiple outputs exist as well. 

So how does the perceptron work?
It is a 3 step process. 

### WORKING OF A PERCEPTRON
#### 1. INPUT APPLICATION
We apply inputs to the perceptron. These inputs are basically numerical values. In case of categorical variables the inputs are numerical values defining the existance of a category. 

#### 2. NET ACTIVATION CALCULATION
The perceptron sums the inputs by a weights formula to calculate the net activation. The value of the variables (for example X1 = 12) is multiplied by the weight of the variable (for example W1 = 0.6) and summed up. 

So the formula becomes:

    - NET ACTIVATION = (X1 x W1) + (X2 x W2) + (X3 x W3) + (X4 x W4) + (X5 x W5)

    - NET ACTIVATION = SUMMATION ( WEIGHTS * INPUTS ) 

#### 3. ACTIVATION
The value of the NET ACTIVATION is passed to the activation function. The function calculates another numerical value which becomes an INPUT to the NEXT NEURON and the process continues.

    - OUTPUT = ACTIVATION ( NET ACTIVATION )

### BUT...
What is an activation function?

Activation function, as described by it's name, ACTIVATES a neuron. It is simply a function that calculates the value that the neuron shoots at it's output node (hence activation).


## ACTIVATION FUNCTION
The following activation functions are the most common in the world of Data Science. 

### STEP FUNCTION

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-1, 1, 1000)
y1 = np.zeros(500)
y2 = np.ones(500)
y = np.concatenate((y1,y2), axis=0)
plt.plot(x,y, label="Step Function")
plt.legend()
plt.grid()
plt.show()
```

No the code is not important. The function you see above is the step function. It is 0 for the values of x below zero and 1 for the values of x greater than zero. 

####  DEFINITION:
    - ACTIVATED FOR X >= 0
    - NOT ACTIVATED FOR X < 0

Can you guess where we might need the step function? 
At the output node of a binary category classification. See how intuitive this is?

#### DRAWBACKS:
    - Multiple Class Activation problem
    - Dramatic changes

### LINEAR FUNCTION

#### DRAWBACKS:

### SIGMOID FUNCTION

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
x=np.linspace(-10,10,100)
plt.plot(x,sigmoid(x),'r', label='Sigmoid')
plt.legend()
plt.grid()
plt.show()
```

####  DEFINITION:
    - y = 1 / ( 1 + exp(-x) )

#### DRAWBACKS:
    - Near the extremities, the network refuses to learn.
    - Dynamic

### HYPERBOLIC TANGENT

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10, 10, .1)
t = np.tanh(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, t, label='tanh')
ax.set_ylim([-1.2, 1.2])
ax.set_xlim([-10,10])
ax.set_xlabel('x')
ax.legend()
ax.grid()
plt.show()
```

####  DEFINITION:
    - tanh = ( 2 sigmoid (2x) )- 1

#### DRAWBACKS:
    - Near the extremities, the network refuses to learn.
    - Best performance

### RECTIFIED LINEAR UNITS ReLU

## BIAS TERM

Alright wait a minute, so the formula of our sum is: 
    - SUM = Summation ( Weights * Values ) 
Where do our weights come from? 
    - They are randomly initialized by our model.
What happens if the inputs are zero
    - Well the input value will have no effect on the output value. 
Is that bad? 
    - Yes, Yes it is. Because zero multiplied by any weight will always be zero. Hence the update of the weights will not be possible.
Oh, so how do we fix it?
    - Just add a bias term. It is a constant that is added to all units except the input units. 
So the updated formula becomes:
    - NET ACTIVATION = SUMMATION ( WEIGHTS * INPUTS ) + BIAS

## MULTIPLE PERCEPTRONS = NEURAL NETWORKS

If we connect many single layer perceptrons together we will get a neural network. It will mainly have three layers.
- INPUT
    - Real Data
    - The data is processed and ready for learning.
- HIDDEN
    - Layers for learning
    - More layers refer to a deeper network. 
    - More layers don't necessarily mean a better learning experience. 
    - The number of hidden neurons and the number of hidden layers are optimized by a lot of trial and error.
    - Bias added.
- OUTPUT
    - Output estimate.
    - It could be the probability of the output. (Output activation function : Sigmoid).
    - It could be a category. (Output activation function : Step).
    - Bias added.
    

## COST FUNCTION
Cost functions are a quantitative measure of how far we are from the actual or expected value.
### QUADRATIC COST FUNCTION
Quadratic or square cost functions are extremely common as they show large errors quite well. (Square of a large number is a huge number)

Calculation causes a slow learning for the neural network.

#### FORMULA 
    - C = SUMMATION [ ( PREDICTED - ACTUAL ) ^ 2 ] / N
    
### CROSS ENTROPY FUNCTION
This function allows faster learning speed for the neural network. 

#### FORMULA 
Y' = Predicted Value
Y = Actual Value
N = Total Length of Dataset

    - C = ( -1 / N) x SUMMATION ( Y' ln Y + (1-Y') ln (1-Y) )

#### INTUITION
- Minimizing the cost function means we are closest to the estimation. 
- The closer we are to the estimation the better our model is.

## GRADIENT DESCENT
OR THE PROCESS OF LEARNING.

Gradient Descent refers to an optimized algorithm for the calculation of the minimum of the input function. What function might we need to minimize? THE COST FUNCTION. 
The cost function minimzation refers to the best guesses. 

LETS BREAK IT DOWN!
- Calculate the dradient of the input function.
- Find the direction of the steepest descent. 
- Add or Subtract a certain amount from the input in the same direction. 
- Repeat till the value of change becomes extremely small.

### LEARNING RATE
Specifies how quickly we will learn. 
- Small learning rate might take too long to learn.
- Large learning rate might overshoot the minimum of the cost function.

## BACKPROPAGATION

- Calculate the error of the output and distribute via derivatives, through the network.
